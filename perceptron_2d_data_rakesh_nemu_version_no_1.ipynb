```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('annotated_points_np_200.csv')
```

```python
X_train=data.iloc[:,:-1].values
y_train=data.iloc[:,-1].values
```

```python
class Perceptron_zeros:
    def __init__(self, learning_rate=0.01, n_iters=1000):
        self.learning_rate = learning_rate
        self.n_iters = n_iters
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        n_samples, n_features = X.shape

        self.weights = np.zeros(n_features)
        self.bias = 0
        for _ in range(self.n_iters):
            for idx, x_i in enumerate(X):
                linear_output = np.dot(x_i, self.weights) + self.bias
                predicted_label = self.predict_single(x_i)
                if predicted_label != y[idx]:
                    update = self.learning_rate * (y[idx] - predicted_label)
                    self.weights += update * x_i
                    self.bias += update

    def predict_single(self, x):
        linear_output = np.dot(x, self.weights) + self.bias
        return 2 if linear_output >= 0 else 1

    def predict(self, X):
        return [self.predict_single(x) for x in X]
    def plotf(self,X_train,y_train):
        plt.figure(figsize=(15, 10))
        plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap='viridis', marker='o', edgecolors='k')
        xmin, xmax = np.min(X_train[:, 0]), np.max(X_train[:, 0])
        ymin, ymax = np.min(X_train[:, 1]), np.max(X_train[:, 1])
        xx, yy = np.meshgrid(np.linspace(xmin, xmax, 100), np.linspace(ymin, ymax, 100))
        Z = np.array([perceptron.predict_single(np.array([xi, yi])) for xi, yi in zip(xx.ravel(), yy.ravel())])
        Z = Z.reshape(xx.shape)
        plt.contourf(xx, yy, Z, alpha=0.3, levels=[0, 0.5, 1], colors=('blue', 'orange'))
        plt.xlabel('Feature 1')
        plt.ylabel('Feature 2')
        plt.title('Perceptron Classifier Decision Boundary')
        plt.show()



```
